```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
t = ['0:30', '1:00', '1:30', '2:00', '2:30', '3:00', '3:30', '4:00', '4:30', '5:00']
t = list(map(str, t))
```

```python
location1 = [36,37,37,37,37,38,37,37,38,38]
faculty = location1.copy()
```

```python
location2 = [34,35,34,34,33,34,34,32,32,34]
bikeshare = location2.copy()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)
fig.suptitle('Lower Faculty Glade and Street next to Bikeshare Station\n Particulate Matter levels on September 9, 2022', fontsize=20)
plt.plot(t, location1, label="Site A: Lower Faculty Glade", color='b', linewidth=0.5)
plt.plot(t, location2, label="Site B: Street next to bikeshare station", color='r', linewidth=0.5)


plt.xlabel('Time (mm:ss)', fontsize=20)
plt.ylabel('PM2.5 concentration (ug/m^3)', fontsize=20)
plt.grid()
plt.legend(fontsize=10)
plt.savefig("lab3_time_series_plot.png")
plt.show()
```
