### Reading data from a social network survey

Let us start by downloading and installing the package

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt 
import networkx as nx
import pandas as pd
```

### Define Graph and Read the Edges 

```python
g = nx.DiGraph()
```

```python
fl=pd.read_csv('SchoolEdges.csv',delimiter=' ')
fl.columns = list(map(str.lower, fl.columns)) #change to lower case
```

```python
fl.columns
```

```python
fl
```

```python
 for L in fl.index.values:
        g.add_edge(fl['source'][L],fl['target'][L],weight=fl['weight'][L])
```

```python
g.in_degree()
```

```python
g.out_degree()
```

```python
plt.figure(1,figsize=(16,16)) 
nx.draw_networkx(g,node_size=120,font_size=12)
```

### Extract Largest Component

```python
 Gcc = sorted(nx.weakly_connected_components(g), key=len, reverse=True) # identify largest connected component of the network
```

```python
len(Gcc) ## Note we have 2 components
```

```python
Gcc[1]
```

```python
Gcc[0]
```

```python
G = g.subgraph(Gcc[0])
```

```python
plt.figure(1,figsize=(16,16)) 
nx.draw_networkx(G,node_size=120,font_size=12)
```

#### Write the neighbours of each node 

```python
for node in G.nodes():
    print ("Neighbors of ", node, " are : ", list(G.neighbors(node)))
```

```python
lengths = nx.shortest_path_length(G)
```

```python
lengths
```

```python
lengths = nx.shortest_path_length(G)
st_length = {}
maxlength = 0
for key in lengths:
    ll = key
    i = max(ll[1].values())
    st_length[ll[0]] = i
    if i > maxlength:
        maxlength = i
    print("For Node ", ll[0]," the max length is ",i)

print("The maximun length in the Social network is ",maxlength)
```
